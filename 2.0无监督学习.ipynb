# 1.0对比，有监督学习和无监督学习的图


<font size=5>#以上是有监督学习，也就是数据集相当于一个答案集。在数据集中我们已经明确知道哪些是恶性肿瘤，哪些是良性肿瘤</font>


<font size=5>#以上是无监督学习，所以在无监督学习中 我们只有一个数据集 没人告诉我们该怎么做 我们也不知道 每个数据点究竟是什么意思 相反 它只告诉我们 现在有一个数据集 你能在其中找到某种结构吗？</font>



<font size=5>#无监督算法实际上会把这些数据分成两个不同的聚类，所以实际上就是所谓的聚类算法</font>

# 2.0无监督学习的例子（聚类）

    
<font size=5>   2.1如谷歌新闻news.google.com，当你去搜索某个新闻的时候，谷歌会把所有有关的新闻都聚类起来，他们的主题都是相同的。
    在无监督学习中，我们事先给定的数据集中，我们不知道这些数据是些什么东西，也不知道这些数据的类型。但是 请问你可以自动的找到这些数据中的类型吗？ 然后自动的 按得到的类型把这些个体分类 虽然事先我并不知道哪些类型 因为对于这些数据样本来说 我们没有给算法一个 正确答案 所以 这就是无监督学习
    
    总的来说无监督学习，就是在我们没有给“正确答案”情况下，算法自动在现有数据集中挖掘出数据的聚类。例如有一堆的客户联系名单，但是无从知道应该如何划分这些客户，哪些客户属于市场1，哪些属于市场2，无从知道，无监督学习就是自动的从数据中挖掘出这些
</font>

# 3.0无监督学习的例子（鸡尾酒宴会非集群）


<font size=5>无监督学习能够准确的分出两个人重叠的声音，由于两人距离麦克风的距离不同，所以录进去重合的声音也有所区别，但是我们在给定的声音数据集的时候，并没有指定任何的东西，没有任何“正确答案”，不像监督学习那样，某几个属性时对应起来的。在无监督学习中没有，但是算法要从数据集中准确的找到他们聚类的依据，然后把声音分离出来 </font>

# 总结
<font size=5>无监督学习使我们几乎或根本不了解结果应该是什么样子。 我们可以从数据中获得结构，而不必知道变量的影响。
我们可以通过基于数据中变量之间的关系对数据进行聚类来推导此结构。
在无监督学习的情况下，没有基于预测结果的反馈。</font>
