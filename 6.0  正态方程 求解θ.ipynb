<font size=5><br>5.0更一般的假设我们有m条训练集，每条训练集有n个特征，所以每条训练集的特征矩阵看起来应该是和左边那个x[i]一样的，有n+1个数据，其中x[0]=1,x[i]是一个n+1维的向量。
         <br>那么如上图的设计矩阵X（design matrix X）是怎么定义出来的呢。这个X的每一行就是一个特征矩阵的转置，因此可以说每一行，就是一条训练集。因此这个X就是一个mxn+1的矩阵

<font size=5><br>8.0什么时候使用梯度下降，什么时候使用正态方程？
    <br>
    <br>
    <center>梯度下降的缺点 vs 正态方程的优点</center>
    <br>1.0梯度下降需要选择合适的α，并且需要多步迭代
    <br>2.0正态方程不需要选择α，一步就可求出θ
</font>

![title](43.png)


<font size=5>
    <br>
    <br>
    <center>梯度下降的优点 vs 正态方程的缺点</center>
    <br>1.如果特征很多，梯度下降依旧有效
    <br>2.0但是正态方程需要算一个逆矩阵，而这个逆矩阵是nxn的，计算次数大致是n的三次的时间复杂度
    <br>如果这个n达到上万的话 万x万会是一个不小的数字就会开始变慢了 
    <br>3.0梯度下降法使用与多种学习算法求解参数θ，但是正态方程就不是这样一回事了
</font>
