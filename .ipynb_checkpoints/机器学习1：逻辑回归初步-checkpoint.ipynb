# 现实
线性回归能够对连续值结果进行预测，而现实中有会面临中另外一个问题，分类问题。最简单的情况就是。我在进行英雄联盟回归预测，判断是赢是输时，医生判断病人是否有病时。这些就是一个二分类问题。那么怎么去解决这样的二分类问题呢。如当英雄联盟预测，出现50或者90时，那预测是获胜还是失败呢？这样一个二分类问题应该怎么解决呢。

如果我们设置一个阈值，当预测出来的值大于50时，我们就认为这局游戏可以获得胜利。如下图所示


如果hθ(x)是对应的特征线性组合，如果我们设定一个阈值0.5，大于0.5的点我们认为是可以获得胜利的局，低于0.5则认为失败，这就很好的处理这个二分类问题

# 特例
如出现这样一种情况，在回归预测中，存在预测值大于0.5的局，但该局最终的结果是失败了，而小于0.5的值最终有获胜了。如下图所示


绿色点出现了错误，而实际情况下，但但依靠线性回归+阈值的方式，已经很难完成一个鲁棒性（健壮和强壮的意思）很好的分类器了
在这样的情况下，逻辑回归就诞生了


# 逻辑回归
逻辑回归的核心思想，线性回归结果输出的是一个连续值，而这个连续值是无法限定的，那么我们就无法把这个结果映射为可以帮助我们判断的结果。
而如果输出的结果是（0，1）这样的概率值，那么就可以很好的解决分类问题了。数学上有个函数叫做Sigmoid函数



图像如下


可以看出来z=0是该函数的取值为1/2，而z逐渐变小，函数的值就趋于0，z变大就逐渐趋于1.因为只要通过线性回归得出结果后，在把数值放到这个Sigmoid函数，就可以得到一个概率值。这样就可以解决二分类的问题
