# 房价大小与价格的预测
<font size=5><br>1.0首先这是一个监督学习，因为你朋友收集到的纽约房价与房子大小的数据集，里面的每一项都是一个正确的答案，因此这就是一个监督学习</font>
<font size=5><br>2.0再者这是一个回归问题，因为它能够更根据我们之前的数据预测出一个准确的输出值</font>
<font size=5><br>3.0在这里，如果建模是一条一次函数，那么你朋友的房子1250英尺，就应该卖220k一英尺</font>


# 在监督学习中，我将需要一个“正确答案集”，也就是所谓的数据集也叫训练集，下面定义了一些符号

<font size=5>m:训练样本的数目
      <br>x:输入变量/输入特征
      <br>y:输出变量/目标变量
      <br>使用（x,y）来表示一个训练样本，也就是一个条目，（房子大小，房子价格）
      <br>（x[i],y[i]）左上角i 表示第i个条训练样本

</font>


# 机器学习的过程如下

<font size=5>
<br>1.准备训练集
<br>2.训练集输入到学习算法中
<br>3.得到一个函数h（hypothesis假设）
<br>4.你的朋友输入房子的大小x通过函数h
<br>5.得出y，房子的销售价格
</font>

# 接下来我们要思考，有了数据集我们如何得到这个h
<font size=5>我们一般如此表示，最后会得到一条线性函数，当然我们会遇到更复杂的函数不一定是线性的</font>
